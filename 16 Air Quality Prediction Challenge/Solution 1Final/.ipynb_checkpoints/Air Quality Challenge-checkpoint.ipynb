```python
import numpy as np
import pandas as pd 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Train.csv')
df=df.values
print(df)
xtrain=df[:,:5]
ytrain=df[:,5:6]
```

```python
lr=LinearRegression(normalize=True)
print(lr)
```

```python
lr.fit(xtrain,ytrain)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
test=pd.read_csv('Test/Test.csv')
test=test.values
print(test.shape)
```

```python
res=lr.predict(test)
# lr.score(test,res)
print(res)
```

```python
#Use this Statement OR below Commented  Complete Function
pd.DataFrame(res).to_csv("output.csv")
```
